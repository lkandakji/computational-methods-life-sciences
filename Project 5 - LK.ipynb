## Motivation
You will analyse data from [this paper](https://www.journals.uchicago.edu/doi/abs/10.1086/285144) by [Dr. David Currie](https://science.uottawa.ca/biology/people/currie-david). The data (`ENVDATA.csv` in csv format) is provided along this notebook. Here is the paper's abstract:

<span style="font-family: Babas; font-size: 1em;">
Many hypotheses have been proposed to explain the great variation among regions in species richness. These were tested by first examining patterns of species richness of birds, mammals, amphibians, and reptiles in 336 quadrats covering North America. These patterns were then compared with the regional variation of 21 descriptors of the environment suggested by the hypotheses. I found that, in the four vertebrate classes studied, 80%-93% of the variability in species richness could be statistically explained by a monotonically increasing function of a single variable: annual potential evapotranspiration (PET). In contrast, tree richness is more closely related to actual evapotranspiration (AET). Both AET and PET appear to be measures of available environmental energy. The relationships between tree and vertebrate richness are strikingly poor. Species richness in particular orders and families of the Vertebrata is also closely related to PET, but not always monotonically, often resembling a replacement series along an environmental gradient. The present results are consistent with the hypothesis that environmentally available energy limits regional species richness. However, my observations are not completely consistent with earlier species-energy theory. The energy-richness relationship appears to depend on scale, and it is affected differently by variations in area and in areal energy flux.
</span>

## P2

Add a column indicating whether the temperature is above or below the freezing point, 

(i.e,  `df[‘is_above_zero’] = df[‘TMP’] > 0` )

Answer to the following:

    1 - What are the mean values of the numbers of each species within each group (above,below zero)
    2 - What are the variances in number of species?
    3 - Which species has the highest variance? How about mean?


```python
import pandas as pd

def freezing_point(c):
    if c['TMP'] < 0:
        return 'Below'
    elif c['TMP'] > 0:
        return 'Above'
    elif c['TMP'] == 0:
        return 'At FP'
    else:
        return 'Undefined'
    
df = pd.read_csv('ENVDATA.csv')
df['Above/Below FP'] = df.apply(freezing_point, axis=1)
df.to_csv('ENVDATA_Problem2.csv', index=False)

print('1. What are the mean values of the numbers of each species within each group (above and below FP)?')

df_0 = df[df['Above/Below FP'] == 'Below']
df_1 = df[df['Above/Below FP'] == 'Above']

print('')
print('BELOW FREEZING POINT')
print('Average number of breeding bird species: '+ str(df_0['BIRD'].mean()))
print('Average number of woody tree species: ' + str(df_0['TREE'].mean()))
print('Average number of reptile species: '+ str(df_0['REPT'].mean()))

print('')
print('ABOVE FREEZING POINT')
print('Average number of breeding bird species: '+ str(df_1['BIRD'].mean()))
print('Average number of woody tree species: ' + str(df_1['TREE'].mean()))
print('Average number of reptile species: '+ str(df_1['REPT'].mean()))

print('')
print('2. What are the variances in number of species?')
print('')

print('Variance of breeding bird species: '+ str(df.loc[:,"BIRD"].var()))
print('Variance of woody tree species: ' + str(df.loc[:,"TREE"].var()))
print('Variance of reptile species: '+ str(df.loc[:,"REPT"].var()))

print('')
print('3. Which species has the highest variance? How about mean?')
print('')

def highest_var(df):
    b = df.loc[:,"BIRD"].var()
    t = df.loc[:,"TREE"].var()
    r = df.loc[:,"REPT"].var()
    Max = b
    if t > Max:
        Max = t    
    if r > Max:
        Max = r
        if t > r:
            Max = t
    if Max == b:
        return('breeding bird species, ' + str(Max))
    if Max == t:
        return('woody tree species, ' + str(Max))
    if Max == r:
        return('reptile species, ' + str(Max))

def highest_mean(df):
    b = df['BIRD'].mean()
    t = df['TREE'].mean()
    r = df['REPT'].mean()
    Max = b
    if t > Max:
        Max = t    
    if r > Max:
        Max = r
        if t > r:
            Max = t
    if Max == b:
        return('breeding bird species, ' + str(Max))
    if Max == t:
        return('woody tree species, ' + str(Max))
    if Max == r:
        return('reptile species, ' + str(Max))

print('Species with highest variance: ' + str(highest_var(df)))
print('Species with highest mean: ' + str(highest_mean(df)))
```

```python
print('2. ')

corr = df.corr()
plt.figure(figsize = (16,10))
sb.heatmap(corr, fmt="g", cmap='RdBu')
plt.title('CORRELATION HEATMAP')
plt.show()
```

```python
def get_redundant_pairs(df):
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_highest_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("3. Highest Correlations: ")
print('')
print(get_highest_correlations(df, 24))
```
