## Problem 1
    1- Write a function that selects a number at random from a list of integers and if that number is even it replaces it by its half (n/2) and if it is odd it does not do anything.
    2 - Try this function on [10,-3,30,15,2,20] 20 times and print the output each time (better to use a for loop)
    3 - What happens if you try this 1000 times?
    
NOTE: Make sure that you take care of all possible edge cases


```python
from random import *

def randomNumberFromList(n):
    listlength = len(n)
    if (listlength>0):
        randIndex = randint(0,listlength-1)
        if (n[randIndex]%2 == 0):
            n[randIndex]/=2
        print(n)
        
print("Trying the function 20 times:")

for i in range(20):
    randomNumberFromList([10,-3,30,15,2,20])

print(" ")
print("Trying the function 1000 times:")

for i in range(1000):
    randomNumberFromList([10,-3,30,15,2,20])
```

## Problem 2
    1 - Write a function called random_coin_flips that takes an integer as input and prints a string sample of n coin flip trials, in addition it 'returns' the number of heads in the outcome. For example if you run this function as x = random_coin_flips(7), then you should see something like 'HTTHHTH' and x should be equal 4.
    
    2 - Run the function random_coin_flips for n= 100, 500 times and store the output in an array (keep in mind that the array should only contain the number of heads in each run and not the run of coin flips)
    
    3 - Plot the histogram of the data obtained in part 2.

```python
import random

def random_coin_flips(n):
    flips = ['H','T']
    sequenceString=""
    for i in range(n):
        sequenceString+=random.choice(flips)
    print(sequenceString)
    return sequenceString.count('H')

results = []
print("100 flips: ")
results.append(random_coin_flips(500))

print(" ")
print("List of Results: ")
print(results)
```

## Problem 3
A simple random walk, is a random process in one dimension. Imagine that the black dot in the following image can only move one step to the left (i.e. current location -1) or one step to the right (i.e. current location + 1) at random with equal probabilities (p = 0.5).
It is as if it flips a coin and if it comes up head it goes to the right, and if it comes up tail it goes to the left.
<img src="RW1.png" width="500" height="600">
The prorcess that describes the distance of the the black dot from its origin at a given step is a simple ranodm walk. See [this link](http://www.mit.edu/~kardar/teaching/projects/chemotaxis%28AndreaSchmidt%29/random.htm) for more information.

    1 - Write a function called simple_random_walk that simulates a random walk starting from 0. It takes as an input the initial position of the point (in this case 0), and the total number of steps (N), and outputs the path (i.e. position of the point at every step up to step N).
    HINT: you can start with a list x = [0] and then choose whether to go left or right (-1 or 1) at random (e.g. a=np.random.choice([-1,1])). Then add this value to the last element in the list x (i.e. x[-1]) and append it to the list.
    2 - Plot three different realizations of the path.

```python
import numpy as np

def simple_random_walk(initialPosition, N):
    import numpy as np
    import matplotlib.pyplot as plt
    pathList = [initialPosition]
    for i in range(N):
        LeftorRight = np.random.choice([-1,1])
        lastElement = pathList[len(pathList)-1]
        pathList.append(lastElement+LeftorRight)
    print(pathList)
    plt.plot(pathList)
    return plt.show()

print('One realization of the path: ')
simple_random_walk(0,4)
print('Another realization of the path: ')
simple_random_walk(0,4)
print('One more realization of the path: ')
simple_random_walk(0,4)
```

## Problem 4
Another method in `pyplot` is `plt.bar`, which draws bar plots. Using the documentation for bar plots in matplot lib, write a function that takes a DNA string and plots nucleotide frequencies as a bar plot. For example, if you run your function on <font color='red'> gattcaacta </font> it shows the following plot:
<img src="P4.png" width="500" height="600">
Apply your function to the following sequence and plot the frequencies.

```python
def frequencyA(my_seq):
    myseq = my_seq.lower()
    countA = my_seq.count('a')
    numofbases = len(my_seq)
    return float(countA)/numofbases
    
def frequencyT(my_seq):
    myseq = my_seq.lower()
    countT = my_seq.count('t')
    numofbases = len(my_seq)
    return float(countT)/numofbases

def frequencyC(my_seq):
    myseq = my_seq.lower()
    countC = my_seq.count('c')
    numofbases = len(my_seq)
    return float(countC)/numofbases

def frequencyG(my_seq):
    myseq = my_seq.lower()
    countG = my_seq.count('g')
    numofbases = len(my_seq)
    return float(countG)/numofbases

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = ['A', 'T', 'C', 'G']
frequency = [frequencyA(my_seq), frequencyT(my_seq), frequencyC(my_seq), frequencyG(my_seq)]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, frequency, color='green')
plt.xlabel("Nucleotide Base")
plt.ylabel("Frequency of Occurence")
plt.title("Frequency of Nucleotides in a DNA Sequence")

plt.xticks(x_pos, x)

plt.show()

my_seq='cgctcttgtcgccactgcaggagaccgggtcctttttgttatcctccttggcggtcacccctccgtggcgcgggggatacgccccccgaggcgagttctt'
```
